**Produce a table with the average gross earning and budget (`gross` and `budget`) by genre. Calculate a variable `return_on_budget` which shows how many \$ did a movie make at the box office for each \$ of its budget. Ranked genres by this `return_on_budget` in descending order**

```python
# Sorting movies by average gross and budget within each genre, and calculating return on budget
genre_gross_budget = movies.groupby('genre')[['gross', 'budget']].mean()
genre_gross_budget['return_on_budget'] = genre_gross_budget['gross'] / genre_gross_budget['budget']
genre_gross_budget = genre_gross_budget.sort_values(by='return_on_budget', ascending=False)
genre_gross_budget[['gross', 'budget', 'return_on_budget']]
```

**Produce a table that shows the top 15 directors who have created the highest gross revenue in the box office. Don't just show the total gross amount, but also the mean, median, and standard deviation per director**

```python
# Aggregate gross statistics by director
# Display the top 15 directors with their total, mean, median, and std of gross revenue
director_gross_stats = movies.groupby('director')['gross'].agg(
    total_gross='sum',
    mean_gross='mean',
    median_gross='median',
    std_gross='std'
).sort_values(by='total_gross', ascending=False).head(15)

director_gross_stats
```

**Examine the relationship between `gross` and `rating`. Produce a scatterplot, faceted by `genre` and discuss whether IMDB ratings are likely to be a good predictor of how much money a movie will make at the box office. Is there anything strange in this dataset?**

```python
# Relationship between IMDB rating and gross revenue by genre
# Scatter plot to visualize the relationship between IMDB rating and gross revenue, colored by genre
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies, x='rating', y='gross', hue='genre', alpha=0.7)
plt.xlabel('IMDB Rating')
plt.ylabel('Gross Revenue (USD)')
plt.title('Relationship between IMDB Rating and Gross Revenue by Genre')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Challenge - Produce a visualisation of your choosing, something you think is important to explore/investigate visually. Do make it a comprehensive graph and not a single histogram or density plot. Make it publication ready, label your axes, give it a title, choose colouring options, appropirately formatted text etc.**

```python
# Visualisation: Genre-wise Distribution of IMDB Ratings and Gross Revenue

# Prepare data for plotting
genre_stats = (
    movies.groupby('genre')
    .agg(
        avg_rating=('rating', 'mean'),
        avg_gross=('gross', 'mean'),
        count=('title', 'count')
    )
    .reset_index()
)

# Sort genres by average rating for better visual order
genre_stats = genre_stats.sort_values(by='avg_rating', ascending=False)

fig, ax1 = plt.subplots(figsize=(13, 7))

# Bar plot: Average IMDB rating by genre
sns.barplot(
    data=genre_stats,
    x='genre',
    y='avg_rating',
    hue='genre',
    palette='viridis',
    ax=ax1
)
ax1.set_ylabel('Average IMDB Rating', fontsize=12)
ax1.set_xlabel('Genre', fontsize=12)
ax1.set_xticks(range(len(genre_stats['genre'])))
ax1.set_xticklabels(genre_stats['genre'], rotation=30, ha='right', fontsize=11)
ax1.tick_params(axis='y', labelsize=11)

# Line plot: Average gross revenue by genre (secondary axis)
ax2 = ax1.twinx()
sns.pointplot(
    data=genre_stats,
    x='genre',
    y='avg_gross',
    color='crimson',
    markers='D',
    linestyles='-',
    ax=ax2
)
ax2.set_ylabel('Average Gross Revenue (USD)', fontsize=12, color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson', labelsize=11)
ax2.set_ylim(0, genre_stats['avg_gross'].max() * 1.2)

# Annotate number of movies per genre above bars
for idx, row in genre_stats.iterrows():
    ax1.text(
        idx, row['avg_rating'] + 0.08, f"{int(row['count'])}", 
        ha='center', va='bottom', fontsize=10, color='black'
    )

# Title and layout
plt.title(
    'Genre-wise Average IMDB Rating and Gross Revenue\n(Number above bars = Number of Movies)',
    fontsize=15, pad=20
)
ax1.grid(axis='y', linestyle='--', alpha=0.5)
fig.tight_layout()
plt.show()
```
